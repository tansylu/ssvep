```python
# Loaded variable 'df' from URI: /Users/tansylu/Documents/ssvep/filter_stats (4).csv
import pandas as pd
df = pd.read_csv(r'/Users/tansylu/Documents/ssvep/filter_stats (4).csv')
```

```python
# identify bottom 10 percent of filter by score
df = df.sort_values(by='Total Similarity Score', ascending=True)
df = df.reset_index(drop=True)
df = df.head(int(len(df)*0.1))
df.to_csv(r'/Users/tansylu/Documents/ssvep/bottom_10_percent.csv', index=False)
```

```python
# plot by layer distr
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Layer')
plt.title('Distribution of Bottom 10% Filters by Layer')
plt.xlabel('Layer')
plt.ylabel('Count')
plt.show()
```
